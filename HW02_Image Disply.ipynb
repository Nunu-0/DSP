```python
import numpy as np 
from scipy.misc import imread, imresize 
import matplotlib.pyplot as plt

img = imread('C:/Users/Nunu/DSP/code/hedgehog.jpg') # 고슴도치 사진을 img에 불러오기
img_tinted =img * [1, 0.8, 0.8] #  [R, G, B] 로 이미지 색 바꾸기

plt.subplot(1, 2, 1) # Row 1줄. Calumn 2줄의 그래프 중 1 번째 그래프 (원본 사진) 
plt.imshow(img) # 1번째 그래프에 원본사진 오게하기

plt.subplot(1, 2, 2) # 2번째 그래프(색이 변한 사진)
plt.imshow(np.uint8(img_tinted)) # 2번째 그래프에 변한 사진 오게하기

plt.show() # 그래프 보이기
```
